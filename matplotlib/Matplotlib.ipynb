```python
import matplotlib.pyplot as plt
import numpy as np
```

# 创建图

## 1 线性图

```python
np.random.seed(42)
x = np.random.randn(30)
plt.plot(x,"r--o")
```

## 2 线条颜色、标记形状和线型
颜色
b 蓝  r 红  w 白 g 绿  k 黑  y 黄
点状
o  *  +  x
连线
： - -- -.

```python
a = np.random.randn(30)
b = np.random.randn(30)
c = np.random.randn(30)
d = np.random.randn(30)
plt.plot(a,"r--o",b,"b-*",c,"g-.+",d,"y:x")
```

## 3.标签与图例

```python
np.random.seed(42)
x = np.random.rand(30)
y = np.random.rand(30)
# 标题
plt.title("Example")
# 坐标
plt.xlabel("X")
plt.ylabel("Y")

x, = plt.plot(x,"r--o")
y, = plt.plot(y,"b-*")
# 图例
plt.legend([x,y],["X","Y"])
```

## 4 子图   subplot

```python
a = np.random.rand(30)
b = np.random.rand(30)
c = np.random.rand(30)
d = np.random.rand(30)

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

A = ax1.plot(a,"r--o")
ax1.set_title("1")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
# ax1.legend([A],["A"])

B = ax2.plot(a,"b-*")
ax2.set_title("2")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
# ax2.legend([B],["B"])

C = ax3.plot(a,"g-.+")
ax3.set_title("3")
ax3.set_xlabel("x")
ax3.set_ylabel("y")
# ax3.legend([C],["C"])

D = ax4.plot(a,"m:x")
ax4.set_title("4")
ax4.set_xlabel("x")
ax4.set_ylabel("y")
# ax4.legend([D],["D"])
```
